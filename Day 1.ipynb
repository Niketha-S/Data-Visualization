```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#ASSINGMENT-1 (Day-1)
#PLOTTING A GRAPH USING LINE PLOT

x=np.arange(0,10)
y=x*x
plt.plot(x, y, 'go--',linewidth=3, markersize=9,color='red')
plt.xlabel('x axis', color='red',fontsize=15)
plt.ylabel('y axis', color='red',fontsize=15)
plt.title('Assingment-1 Day 1 (LINE PLOT)', color='blue')
plt.show()
```

```python
#ASSINGMENT -2 (DAY 1)
# Plotting between Days and sales
x=[1,2,3,4,5,6,7] 
y=[160,150,140,145,175,165,180] 

plt.plot(x, y, 'ro--',linewidth=2, markersize=10,color='green')
plt.xlabel('x axis',fontsize=15 ,color='red')
plt.ylabel('y axis',fontsize=15 ,color='red')
plt.title('Assingment -2(Day 1)', color='red')
plt.show()
```
